# Assignment 

```python
import pandas
```

```python
df = pandas.read_csv('tmdb_5000_movies.csv')
```

```python
df.head(5)
```

```python
df = df[['id','title','overview','genres','vote_average','vote_count','popularity','revenue','runtime']]
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
top_rated = df.sort_values(by='vote_average', ascending=False).head(10)
top_rated[['title', 'vote_average']]
```

```python
top_revenue = df.sort_values(by='revenue', ascending=False).head(10)
top_revenue[['title', 'revenue']]
```

```python
df[df['runtime'] < 60].sort_values(by='vote_average', ascending=False).head(10)[['title', 'runtime', 'vote_average']]
```

```python
average_runtime = df['runtime'].mean()
print(f"Average runtime of movies: {average_runtime:.2f} minutes")
```

```python
min_runtime = df['runtime'].min()
shortest_movie = df[df['runtime'] == min_runtime][['title', 'runtime']]
shortest_movie.head(5)
```

```python
max_runtime = df['runtime'].max()
longest_movie = df[df['runtime'] == max_runtime][['title', 'runtime']]
longest_movie.head(5)
```
